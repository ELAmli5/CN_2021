```python
from facebook_scraper import get_profile
import time
import pandas as pd
```

```python
dfi = pd.read_csv("Youssoufia_Datas5.csv")
```

```python
user_id=[]
for i in dfi.values:
    user_id.append(i[1])
```

```python
c=0
Users_infos=[]
for user in user_id:
    time.sleep(15)
    user1=get_profile(user, cookies="facebook.com_cookies.txt")
    Users_infos.append(user1)
    c=c+1
    if(c==50):
        time.sleep(120)
        c=0
```

```python
df = pd.DataFrame(Users_infos)
df
```

```python
df.to_csv(r'infos.csv',index=False,encoding='utf-8-sig')
```
